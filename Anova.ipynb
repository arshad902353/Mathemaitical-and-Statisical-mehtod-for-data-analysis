A company wants to test if three different training programs have a significant effect
on the performance scores of their data science trainees after 4 weeks.
  Groups (Training Programs):
 1. Program A – Python-focused
 2. Program B – Machine Learning-focused
 3. Program C – Data Visualization-focused
  Goal: Check if the mean performance score is different among the three
training programs.
 Data (Performance Scores out of 10)

 Using the data above, perform a one-way ANOVA to determine if there is a significant
difference in the mean performance scores among the three training programs.
(Fcritical value ​ = 3.89)

```python
# --- Display Data in Table Form in Colab ---

import pandas as pd

# Sample data
program_a = [6, 7, 8, 7, 9]
program_b = [7, 6, 8, 9, 7]
program_c = [8, 9, 7, 9, 8]

# Create DataFrame
data = pd.DataFrame({
    'Program A': program_a,
    'Program B': program_b,
    'Program C': program_c
})

# Display the table
print("=== Students' Performance Scores ===")
display(data)
```

```python

import numpy as np
from scipy.stats import f

program_a = np.array([6, 7, 8, 7, 9])
program_b = np.array([7, 6, 8, 9, 7])
program_c = np.array([8, 9, 7, 9, 8])

all_data = np.concatenate([program_a, program_b, program_c])

k = 3
n = len(program_a)
N = len(all_data)

mean_a = np.mean(program_a)
mean_b = np.mean(program_b)
mean_c = np.mean(program_c)
grand_mean = np.mean(all_data)

SSB = n * ((mean_a - grand_mean)**2 + (mean_b - grand_mean)**2 + (mean_c - grand_mean)**2)
SSW = (np.sum((program_a - mean_a)**2) + np.sum((program_b - mean_b)**2) + np.sum((program_c - mean_c)**2))

df_between = k - 1
df_within = N - k

MSB = SSB / df_between
MSW = SSW / df_within

F_value = MSB / MSW
p_value = 1 - f.cdf(F_value, df_between, df_within)

print("========== ANOVA SUMMARY ==========\n")
print(f"Number of Groups (k): {k}")
print(f"Sample Size per Group (n): {n}")
print(f"Total Observations (N): {N}\n")

print("---- Group Means ----")
print(f"Program A Mean: {mean_a:.2f}")
print(f"Program B Mean: {mean_b:.2f}")
print(f"Program C Mean: {mean_c:.2f}")
print(f"Grand Mean: {grand_mean:.2f}\n")

print("---- ANOVA Table ----")
print(f"{'Source':<20}{'SS':<15}{'df':<10}{'MS':<15}{'F':<10}{'P-value'}")
print("-" * 70)
print(f"{'Between Groups':<20}{SSB:<15.4f}{df_between:<10}{MSB:<15.4f}{F_value:<10.4f}{p_value:.4f}")
print(f"{'Within Groups':<20}{SSW:<15.4f}{df_within:<10}{MSW:<15.4f}")
print("-" * 70)
print(f"{'Total':<20}{(SSB+SSW):<15.4f}{(df_between+df_within):<10}\n")

alpha = 0.05
if p_value < alpha:
    print("✅ Reject the Null Hypothesis: There is a significant difference among group means.")
else:
    print("❌ Fail to Reject the Null Hypothesis: No significant difference among group means.")

```
